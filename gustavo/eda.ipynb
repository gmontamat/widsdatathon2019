# Exploratory Data Analysis

## Import modules and load data

```python
import matplotlib.pyplot as plt
import os
import pickle

from matplotlib.pyplot import imshow

%matplotlib inline
```

```python
data_path = os.path.join(os.getcwd(), '..', 'input')
```

```python
with open(os.path.join(data_path, 'train_images_256x256.pkl'), 'rb') as fin:
    images = pickle.load(fin)
with open(os.path.join(data_path, 'train_responses.pkl'), 'rb') as fin:
    responses = pickle.load(fin)
with open(os.path.join(data_path, 'train_scores.pkl'), 'rb') as fin:
    scores = pickle.load(fin)
```

## Show positive images

```python
positive = images[responses.squeeze() == 1., :, :, :]
```

```python
for i in range(5):
    plt.figure()
    imshow(positive[i, :, :, :])
```

## Show negative images

```python
negative = images[responses.squeeze() == 0., :, :, :]
```

```python
for i in range(5):
    plt.figure()
    imshow(negative[i, :, :, :])
```

## Some histograms of responses and scores

```python
plt.hist(responses, bins=3)
plt.title("Class imbalance")
plt.show()
```

```python
plt.hist(scores, bins='auto')
plt.title("Distribution of scores")
plt.show()
```

## Show "controversial" images

```python
controversial = images[scores.squeeze() < 0.5, :, :, :]
```

```python
plt.figure()
imshow(controversial[0,:,:,:])
plt.figure()
imshow(controversial[10,:,:,:])
```

## How are controversial responses distributed?

```python
controversial_responses = responses[scores.squeeze() < 0.5, :]
```

```python
plt.hist(controversial_responses, bins=3)
plt.title("Class imbalance of controversial imgs")
plt.show()
```
